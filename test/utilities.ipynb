```python
import pandas as pd

df = pd.read_csv("../top10k_mastodon_users.csv")

with open("users.txt", "w") as f:
    users = df["handle"].tolist()
    f.write("\n".join(users))
```

```python
# Drop the database, don't do on accident!
import chromadb
DB_PATH = "./my_mastodon_db"
COLLECTION_NAME = "mastodon_posts"
client = chromadb.PersistentClient(path=DB_PATH)
verify = input(f"Are you sure you want to drop the database at {DB_PATH}? This action cannot be undone! Type 'YES' to confirm: ")
if verify == "YES":
    client.delete_collection(name=COLLECTION_NAME)
    print("Database dropped.")
```
